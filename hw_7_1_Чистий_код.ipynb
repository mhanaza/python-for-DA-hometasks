````python
from math import pi #Імпорт тільки потрібних модулів
def calc_area(r): #Використання відступу всередині функцій та нижнього підкреслення в іменах функцій
    return pi * r ** 2 #Додала пробіли навколо бінарного оператора
#Додала два порожні рядки між визначеннями функцій calc_area і calc_circumference


def calc_circumference(r): #Використання відступу всередині функцій та нижнього підкреслення в іменах функцій
    return 2 * pi * r #Додала пробіли навколо бінарного оператора

radius = 5 #Додала пробіли навколо бінарного оператора та порожній рядок

print("Area:", calc_area(radius), "Circumference:", calc_circumference(radius)) # Пробіли після коми
````

````python
def CalculateVariance(nl):
    sum_list = 0
    for n in nl:
        sum_list = sum_list + n
    mean = sum_list / len(nl)
    sum_squares = 0
    for n in nl:
        sum_squares = sum_squares + n**2
    mean_squares = sum_squares / len(nl)
    return mean_squares
````

````python
input_list = [75, 80, 85, 90, 95]
````

````python
def calculate_variance(numbers_list):
    """Обчислення дисперсії для списку чисел."""
    # Обчислення середнього значення
    mean = sum(numbers_list) / len(numbers_list)
    
    # Обчислення суми квадратів відхилень від середнього
    sum_squares = sum((n - mean) ** 2 for n in numbers_list)
    
    # Обчислення дисперсії
    variance = sum_squares / len(numbers_list)
    
    return variance
````

````python
variance = calculate_variance(input_list)
variance
````
